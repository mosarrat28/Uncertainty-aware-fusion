```python
import numpy as np
import nibabel as nib
import os
import matplotlib.pyplot as plt
from sklearn.metrics import jaccard_score

# Paths
test_images_path = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_raw/Dataset114_MNMs/imagesTs"
ground_truth_path = "/content/drive/MyDrive/Thesis-ontariotech/DatasetACDC_nnunet/nnUNet_raw/Dataset114_MNMs/labelsTr"
predictions_path = "/content/drive/MyDrive/Thesis-ontariotech/DatasetACDC_nnunet/nnUNet_results/inference_output"

# Label classes (excluding background)
class_labels = [1, 2, 3]
num_classes = len(class_labels)

# Initialize lists to store metrics
per_class_iou = {lbl: [] for lbl in class_labels}
per_class_dice = {lbl: [] for lbl in class_labels}

# Dice function
def dice_coefficient(y_true, y_pred):
    intersection = np.sum(y_true * y_pred)
    return (2. * intersection) / (np.sum(y_true) + np.sum(y_pred) + 1e-8)

# Loop through predictions
for pred_file in sorted(os.listdir(predictions_path)):
    pred_path = os.path.join(predictions_path, pred_file)
    gt_path = os.path.join(ground_truth_path, pred_file)

    if not os.path.exists(gt_path):
        print(f"⚠️ Skipping {pred_file}: Ground truth not found.")
        continue

    pred = nib.load(pred_path).get_fdata().astype(np.uint8)
    gt = nib.load(gt_path).get_fdata().astype(np.uint8)

    for lbl in class_labels:
        pred_bin = (pred == lbl).astype(np.uint8)
        gt_bin = (gt == lbl).astype(np.uint8)

        if gt_bin.sum() == 0 and pred_bin.sum() == 0:
            iou = 1.0  # Perfect match for missing class
            dice = 1.0
        elif gt_bin.sum() == 0 or pred_bin.sum() == 0:
            iou = 0.0
            dice = 0.0
        else:
            iou = jaccard_score(gt_bin.flatten(), pred_bin.flatten())
            dice = dice_coefficient(gt_bin, pred_bin)

        per_class_iou[lbl].append(iou)
        per_class_dice[lbl].append(dice)

# Compute and print results
print("\n📊 Mean Metrics:")
mean_iou_all = []
mean_dice_all = []

for lbl in class_labels:
    mean_iou = np.mean(per_class_iou[lbl])
    mean_dice = np.mean(per_class_dice[lbl])
    mean_iou_all.append(mean_iou)
    mean_dice_all.append(mean_dice)
    print(f"Class {lbl} → IoU: {mean_iou:.4f}, Dice: {mean_dice:.4f}")

print(f"\n Overall Mean IoU: {np.mean(mean_iou_all):.4f}")
print(f"Overall Mean Dice: {np.mean(mean_dice_all):.4f}")
```

```python
import os
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import random

from sklearn.metrics import jaccard_score

# --- Paths ---
test_images_path = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_raw/Dataset114_MNMs/imagesTs"
ground_truth_path = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_raw/Dataset114_MNMs/labelsTr"
predictions_path = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_results/inference_output"

# --- Classes to evaluate ---
labels = [1, 2, 3]  # LV, RV, Myo

# --- Metrics ---
def dice_coefficient(y_true, y_pred):
    intersection = np.logical_and(y_true, y_pred).sum()
    return 2 * intersection / (y_true.sum() + y_pred.sum() + 1e-8)

def iou_score(y_true, y_pred):
    intersection = np.logical_and(y_true, y_pred).sum()
    union = np.logical_or(y_true, y_pred).sum()
    return intersection / (union + 1e-8)

# --- Visualization Function ---
def plot_case(image, gt, pred, slice_idx, case_id):
    gt_masked = np.ma.masked_where(gt == 0, gt)
    pred_masked = np.ma.masked_where(pred == 0, pred)

    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    fig.suptitle(f"Case: {case_id}, Slice: {slice_idx}", fontsize=16)

    # Original
    axes[0].imshow(image, cmap="gray")
    axes[0].set_title("Original Image")
    axes[0].axis("off")

    # Ground Truth Overlay
    axes[1].imshow(image, cmap="gray")
    axes[1].imshow(gt_masked, cmap="jet", alpha=0.5)
    axes[1].set_title("Ground Truth Overlay")
    axes[1].axis("off")

    # Predicted Mask Overlay
    axes[2].imshow(image, cmap="gray")
    axes[2].imshow(pred_masked, cmap="jet", alpha=0.5)
    axes[2].set_title("Predicted Mask Overlay")
    axes[2].axis("off")

    plt.tight_layout()
    plt.show()

# --- Load and Visualize 10 Random Cases ---
test_files = sorted(os.listdir(test_images_path))
random_files = random.sample([f for f in test_files if f.endswith("_0000.nii.gz")], 10)

for image_file in random_files:
    image_file_fixed = image_file.replace("_0000.nii.gz", ".nii.gz")
    case_id = image_file_fixed.replace(".nii.gz", "")

    image_path = os.path.join(test_images_path, image_file)
    pred_path = os.path.join(predictions_path, image_file_fixed)
    gt_path = os.path.join(ground_truth_path, image_file_fixed)

    if not os.path.exists(pred_path) or not os.path.exists(gt_path):
        print(f"⚠️ Skipping {case_id}: missing prediction or GT.")
        continue

    image = nib.load(image_path).get_fdata()
    pred = nib.load(pred_path).get_fdata()
    gt = nib.load(gt_path).get_fdata()

    slice_idx = image.shape[2] // 2
    img_slice = image[:, :, slice_idx]
    gt_slice = gt[:, :, slice_idx]
    pred_slice = pred[:, :, slice_idx]

    # Plot overlays
    plot_case(img_slice, gt_slice, pred_slice, slice_idx, case_id)

    # Compute per-class metrics
    print(f"📊 Metrics for {case_id}")
    for lbl in labels:
        gt_bin = (gt == lbl).astype(np.uint8)
        pred_bin = (pred == lbl).astype(np.uint8)

        dsc = dice_coefficient(gt_bin, pred_bin)
        iou = iou_score(gt_bin, pred_bin)

        print(f"  Class {lbl}: Dice = {dsc:.4f}, IoU = {iou:.4f}")
    print("-" * 50)
```

```python
import os
import shutil
import subprocess
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import jaccard_score, f1_score

# ===== CONFIG =====
trained_model_dir = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/mnmchallengefinetune/nnUNet_dataset/nnUNet_results/Dataset114_MNMs/nnUNetTrainer__nnUNetPlans__3d_fullres"
checkpoint_name = "checkpoint_best.pth"

acdc_imagesTs = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_raw/Dataset114_ACDC/imagesTs"
acdc_labelsTr = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_raw/Dataset114_ACDC/labelsTr"

temp_test_dir = "/content/tmp_acdc_test/imagesTs"
temp_output_dir = "/content/tmp_acdc_predictions"

# Number of test cases to run
num_cases = 3

# ===== PREP TEMP FOLDERS =====
for p in [temp_test_dir, temp_output_dir]:
    if os.path.exists(p):
        shutil.rmtree(p)
    os.makedirs(p, exist_ok=True)

# Select first N cases
selected_files = sorted([f for f in os.listdir(acdc_imagesTs) if f.endswith("_0000.nii.gz")])[:num_cases]
for f in selected_files:
    shutil.copy(os.path.join(acdc_imagesTs, f), os.path.join(temp_test_dir, f))

print(f"✅ Selected {len(selected_files)} cases for testing:")
for f in selected_files:
    print("   -", f)

# ===== RUN PREDICTION =====
print("\n🚀 Running nnU-Net prediction...")
subprocess.run([
    "nnUNetv2_predict",
    "-i", temp_test_dir,
    "-o", temp_output_dir,
    "--trained_model_dir", trained_model_dir,
    "-f", "0",
    "-chk", checkpoint_name
], check=True)

# ===== EVALUATE + VISUALIZE =====
class_names = {0: "Background", 1: "RV", 2: "MYO", 3: "LV"}

for fname in selected_files:
    case_id = fname.replace("_0000.nii.gz", "")
    pred_path = os.path.join(temp_output_dir, case_id + ".nii.gz")
    label_path = os.path.join(acdc_labelsTr, case_id + ".nii.gz")
    image_path = os.path.join(acdc_imagesTs, fname)

    pred = nib.load(pred_path).get_fdata().astype(np.uint8)
    gt = nib.load(label_path).get_fdata().astype(np.uint8)
    img = nib.load(image_path).get_fdata()

    # Compute macro IoU & Dice
    pred_flat = pred.flatten()
    gt_flat = gt.flatten()
    iou_macro = jaccard_score(gt_flat, pred_flat, average="macro", zero_division=0)
    dice_macro = f1_score(gt_flat, pred_flat, average="macro", zero_division=0)

    # Per-class metrics
    per_class_iou = {}
    per_class_dice = {}
    for c in np.unique(gt):
        gt_c = (gt == c).astype(np.uint8).flatten()
        pred_c = (pred == c).astype(np.uint8).flatten()
        per_class_iou[c] = jaccard_score(gt_c, pred_c, average="binary", zero_division=0)
        per_class_dice[c] = f1_score(gt_c, pred_c, average="binary", zero_division=0)

    print(f"\n📂 Case: {case_id}")
    print(f"   IoU (macro): {iou_macro:.4f}")
    print(f"   Dice (macro): {dice_macro:.4f}")
    for c in per_class_iou.keys():
        print(f"   {class_names.get(c, f'Class {c}')} - IoU: {per_class_iou[c]:.4f}, Dice: {per_class_dice[c]:.4f}")

    # Visualization (middle slice)
    mid_slice = img.shape[-1] // 2
    fig, axs = plt.subplots(1, 3, figsize=(12, 4))
    axs[0].imshow(img[..., mid_slice].T, cmap="gray", origin="lower")
    axs[0].set_title("Image")
    axs[1].imshow(gt[..., mid_slice].T, cmap="gray", origin="lower")
    axs[1].set_title("Ground Truth")
    axs[2].imshow(pred[..., mid_slice].T, cmap="gray", origin="lower")
    axs[2].set_title("Prediction")
    for ax in axs:
        ax.axis("off")
    plt.show()
```

```python
import os
import torch
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import torch.nn.functional as F
from sklearn.metrics import jaccard_score, f1_score

# ===== CONFIG =====
checkpoint_path = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/mnmchallengefinetune/nnUNet_dataset/nnUNet_results/Dataset114_MNMs/nnUNetTrainer__nnUNetPlans__3d_fullres/fold_0/checkpoint_best.pth"
acdc_imagesTs = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_raw/Dataset114_ACDC/imagesTs"
acdc_labelsTr = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_raw/Dataset114_ACDC/labelsTr"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
num_cases = 3  # number of ACDC cases to test
resize_shape = (256, 256)  # resize slices before inference

# ===== LOAD MODEL =====
print("Loading model...")
checkpoint = torch.load(checkpoint_path, map_location=device, weights_only=False)

if isinstance(checkpoint, dict) and 'network' in checkpoint:
    model = checkpoint['network']

elif isinstance(checkpoint, dict) and 'state_dict' in checkpoint:
    from nnunetv2.nets.nets_factory import get_network_from_plans
    plans = checkpoint['plans']
    model = get_network_from_plans(plans, dataset_json=None)
    model.load_state_dict(checkpoint['state_dict'])

else:
    # Assume full nnU-Net Trainer object
    trainer = checkpoint
    model = trainer.network

model.to(device)
model.eval()

# ===== SELECT CASES =====
files = sorted([f for f in os.listdir(acdc_imagesTs) if f.endswith("_0000.nii.gz")])[:num_cases]
class_names = {0: "Background", 1: "RV", 2: "MYO", 3: "LV"}

# ===== INFERENCE LOOP =====
for fname in files:
    case_id = fname.replace("_0000.nii.gz", "")
    image_path = os.path.join(acdc_imagesTs, fname)
    label_path = os.path.join(acdc_labelsTr, case_id + ".nii.gz")

    img_nii = nib.load(image_path)
    lbl_nii = nib.load(label_path)
    img = img_nii.get_fdata().astype(np.float32)
    lbl = lbl_nii.get_fdata().astype(np.uint8)

    # Normalize image (z-score)
    img = (img - np.mean(img)) / (np.std(img) + 1e-8)

    pred_volume = np.zeros_like(lbl, dtype=np.uint8)

    with torch.no_grad():
        for z in range(img.shape[-1]):
            slice_img = img[..., z]
            slice_img_t = torch.from_numpy(slice_img).unsqueeze(0).unsqueeze(0).float().to(device)

            # Resize to expected input size
            slice_img_t = F.interpolate(slice_img_t, size=resize_shape, mode="bilinear", align_corners=False)

            logits = model(slice_img_t)
            probs = torch.softmax(logits, dim=1)
            pred_slice = torch.argmax(probs, dim=1).squeeze(0).cpu().numpy()

            # Resize prediction back to original size
            pred_slice_resized = F.interpolate(
                torch.from_numpy(pred_slice).unsqueeze(0).unsqueeze(0).float(),
                size=slice_img.shape,
                mode="nearest"
            ).squeeze().numpy().astype(np.uint8)

            pred_volume[..., z] = pred_slice_resized

    # ===== METRICS =====
    pred_flat = pred_volume.flatten()
    lbl_flat = lbl.flatten()
    iou_macro = jaccard_score(lbl_flat, pred_flat, average="macro", zero_division=0)
    dice_macro = f1_score(lbl_flat, pred_flat, average="macro", zero_division=0)

    print(f"\n📂 Case: {case_id}")
    print(f"   IoU (macro): {iou_macro:.4f}")
    print(f"   Dice (macro): {dice_macro:.4f}")
    for c in np.unique(lbl):
        gt_c = (lbl == c).astype(np.uint8).flatten()
        pred_c = (pred_volume == c).astype(np.uint8).flatten()
        iou_c = jaccard_score(gt_c, pred_c, average="binary", zero_division=0)
        dice_c = f1_score(gt_c, pred_c, average="binary", zero_division=0)
        print(f"   {class_names.get(c, f'Class {c}')} - IoU: {iou_c:.4f}, Dice: {dice_c:.4f}")

    # ===== VISUALIZATION =====
    mid_slice = img.shape[-1] // 2
    fig, axs = plt.subplots(1, 3, figsize=(12, 4))
    axs[0].imshow(img[..., mid_slice].T, cmap="gray", origin="lower")
    axs[0].set_title("Image")
    axs[1].imshow(lbl[..., mid_slice].T, cmap="gray", origin="lower")
    axs[1].set_title("Ground Truth")
    axs[2].imshow(pred_volume[..., mid_slice].T, cmap="gray", origin="lower")
    axs[2].set_title("Prediction")
    for ax in axs:
        ax.axis("off")
    plt.show()
```

```python
import os
import torch
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import torch.nn.functional as F
from sklearn.metrics import jaccard_score, f1_score
from nnunetv2.training.nnUNetTrainer.nnUNetTrainer import nnUNetTrainer

# ===== CONFIG =====
checkpoint_path = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/mnmchallengefinetune/nnUNet_dataset/nnUNet_results/Dataset114_MNMs/nnUNetTrainer__nnUNetPlans__3d_fullres/fold_0/checkpoint_best.pth"
acdc_imagesTs = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_raw/Dataset114_ACDC/imagesTs"
acdc_labelsTr = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_raw/Dataset114_ACDC/labelsTr"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
num_cases = 3  # how many test cases from ACDC to evaluate
resize_shape = (256, 256)  # resize for inference

# ===== LOAD MODEL =====
print("Loading model...")
checkpoint = torch.load(checkpoint_path, map_location=device, weights_only=False)

if 'network_weights' in checkpoint:
    init_args = checkpoint['init_args']
    trainer = nnUNetTrainer(plans=init_args['plans'],
                            configuration=init_args['configuration'],
                            fold=init_args['fold'],
                            dataset_json=init_args['dataset_json'],
                            device=init_args['device'])

    model = trainer.build_network_architecture(
        plans=init_args['plans'],
        dataset_json=init_args['dataset_json'],
        arch_init_kwargs_req_import=init_args.get('arch_init_kwargs_req_import', None),
        num_input_channels=init_args.get('num_input_channels', 1),
        num_output_channel=init_args.get('num_output_channel', len(init_args['dataset_json']['labels']))
    )

    model.load_state_dict(checkpoint['network_weights'])
else:
    raise ValueError(f"❌ Unknown checkpoint format. Keys found: {checkpoint.keys()}")

model.to(device)
model.eval()

# ===== SELECT TEST CASES =====
files = sorted([f for f in os.listdir(acdc_imagesTs) if f.endswith("_0000.nii.gz")])[:num_cases]
class_names = {0: "Background", 1: "RV", 2: "MYO", 3: "LV"}

# ===== INFERENCE LOOP =====
for fname in files:
    case_id = fname.replace("_0000.nii.gz", "")
    image_path = os.path.join(acdc_imagesTs, fname)
    label_path = os.path.join(acdc_labelsTr, case_id + ".nii.gz")

    img_nii = nib.load(image_path)
    lbl_nii = nib.load(label_path)
    img = img_nii.get_fdata().astype(np.float32)
    lbl = lbl_nii.get_fdata().astype(np.uint8)

    # Z-score normalization
    img = (img - np.mean(img)) / (np.std(img) + 1e-8)

    pred_volume = np.zeros_like(lbl, dtype=np.uint8)

    with torch.no_grad():
        for z in range(img.shape[-1]):
            slice_img = img[..., z]
            slice_img_t = torch.from_numpy(slice_img).unsqueeze(0).unsqueeze(0).float().to(device)

            # Resize slice
            slice_img_t = F.interpolate(slice_img_t, size=resize_shape, mode="bilinear", align_corners=False)

            logits = model(slice_img_t)
            probs = torch.softmax(logits, dim=1)
            pred_slice = torch.argmax(probs, dim=1).squeeze(0).cpu().numpy()

            # Resize back to original size
            pred_slice_resized = F.interpolate(
                torch.from_numpy(pred_slice).unsqueeze(0).unsqueeze(0).float(),
                size=slice_img.shape,
                mode="nearest"
            ).squeeze().numpy().astype(np.uint8)

            pred_volume[..., z] = pred_slice_resized

    # ===== METRICS =====
    pred_flat = pred_volume.flatten()
    lbl_flat = lbl.flatten()
    iou_macro = jaccard_score(lbl_flat, pred_flat, average="macro", zero_division=0)
    dice_macro = f1_score(lbl_flat, pred_flat, average="macro", zero_division=0)

    print(f"\n📂 Case: {case_id}")
    print(f"   IoU (macro): {iou_macro:.4f}")
    print(f"   Dice (macro): {dice_macro:.4f}")
    for c in np.unique(lbl):
        gt_c = (lbl == c).astype(np.uint8).flatten()
        pred_c = (pred_volume == c).astype(np.uint8).flatten()
        iou_c = jaccard_score(gt_c, pred_c, average="binary", zero_division=0)
        dice_c = f1_score(gt_c, pred_c, average="binary", zero_division=0)
        print(f"   {class_names.get(c, f'Class {c}')} - IoU: {iou_c:.4f}, Dice: {dice_c:.4f}")

    # ===== VISUALIZATION =====
    mid_slice = img.shape[-1] // 2
    fig, axs = plt.subplots(1, 3, figsize=(12, 4))
    axs[0].imshow(img[..., mid_slice].T, cmap="gray", origin="lower")
    axs[0].set_title("Image")
    axs[1].imshow(lbl[..., mid_slice].T, cmap="gray", origin="lower")
    axs[1].set_title("Ground Truth")
    axs[2].imshow(pred_volume[..., mid_slice].T, cmap="gray", origin="lower")
    axs[2].set_title("Prediction")
    for ax in axs:
        ax.axis("off")
    plt.show()
```

```python
import os
import random
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
from sklearn.metrics import jaccard_score, f1_score

# ==== Paths ====
pred_dir = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_raw/nnUNet_results/inference_output_acdc"
gt_dir = "/content/drive/MyDrive/Thesis-ontariotech/Dataset/ACDC_nnunet/nnUNet_raw/Dataset114_ACDC/labelsTr"

# ==== Class mapping (ACDC) ====
class_labels = {
    1: "RV",
    2: "MYO",
    3: "LV"
}

# ==== MNMs → ACDC label mapping ====
# MNMs: LVBP=1, LVM=2, RV=3
# ACDC: RV=1, MYO=2, LV=3
mnm_to_acdc = {0: 0, 1: 3, 2: 2, 3: 1}

# ==== Metric functions ====
def dice_score(y_true, y_pred, label):
    y_true_bin = (y_true == label).astype(np.uint8).ravel()
    y_pred_bin = (y_pred == label).astype(np.uint8).ravel()
    return f1_score(y_true_bin, y_pred_bin)

def iou_score(y_true, y_pred, label):
    y_true_bin = (y_true == label).astype(np.uint8).ravel()
    y_pred_bin = (y_pred == label).astype(np.uint8).ravel()
    return jaccard_score(y_true_bin, y_pred_bin)

# ==== Evaluation ====
all_dice = {label: [] for label in class_labels}
all_iou = {label: [] for label in class_labels}

common_cases = sorted([f for f in os.listdir(pred_dir) if f.endswith(".nii.gz")])

for file in common_cases:
    pred_path = os.path.join(pred_dir, file)
    gt_path = os.path.join(gt_dir, file)

    if not os.path.exists(gt_path):
        continue

    pred = nib.load(pred_path).get_fdata().astype(np.uint8)
    gt = nib.load(gt_path).get_fdata().astype(np.uint8)

    # Apply label remapping
    pred_remapped = np.zeros_like(pred)
    for src, tgt in mnm_to_acdc.items():
        pred_remapped[pred == src] = tgt

    for label in class_labels:
        all_dice[label].append(dice_score(gt, pred_remapped, label))
        all_iou[label].append(iou_score(gt, pred_remapped, label))

# ==== Print results ====
print("\n=== ACDC Cross-Dataset Evaluation (MNMs-trained nnU-Net, remapped labels) ===")
for label, name in class_labels.items():
    print(f"{name} - Dice: {np.mean(all_dice[label]):.4f}, IoU: {np.mean(all_iou[label]):.4f}")
print(f"Overall Dice: {np.mean([np.mean(v) for v in all_dice.values()]):.4f}")
print(f"Overall IoU: {np.mean([np.mean(v) for v in all_iou.values()]):.4f}")

# ==== Visualization ====
sample_cases = random.sample(common_cases, min(5, len(common_cases)))
for case in sample_cases:
    pred = nib.load(os.path.join(pred_dir, case)).get_fdata().astype(np.uint8)
    gt = nib.load(os.path.join(gt_dir, case)).get_fdata().astype(np.uint8)

    # Apply remapping for visualization
    pred_remapped = np.zeros_like(pred)
    for src, tgt in mnm_to_acdc.items():
        pred_remapped[pred == src] = tgt

    mid_slice = pred.shape[2] // 2
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(gt[:, :, mid_slice], cmap="gray")
    plt.title(f"GT - {case}")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(pred_remapped[:, :, mid_slice], cmap="gray")
    plt.title(f"Prediction (Remapped) - {case}")
    plt.axis("off")
    plt.show()
```
